```python
# 这里是一个模型的实例, 可以调整不同的参数多跑跑看
model = TimeVaryingMeanFieldChiralInducedPhaseLag(
    strengthLambda=19.2,              # 耦合强度 K
    distanceD0=1,                     # 耦合距离 d0
    boundaryLength=7,                 # 边界长度 L
    phaseLagFreqOmega=1,              # 阻锉 alpha0 的频率 omega
    phaseLagAmplitude=np.pi / 2,      # 阻锉 alpha0 的振幅 
    distribution="uniform",           # 自然频率分布
    initPhaseTheta=np.zeros(1000),    # 初始相位
    omegaMin=0.1,                     # 自然频率下限 omega_min 
    deltaOmega=1,                     # 自然频率差 delta_omega
    agentsNum=1000,                   # 振子数 N
    savePath=SAVE_PATH,               # 保存路径
    dt=0.01,                          # 时间步长 dt
    tqdm=True,                        # 是否显示进度条
    overWrite=True,                   # 是否覆盖已有数据
    randomSeed=randomSeed,            # 随机种子(用于获取不同的随机初值)
    shotsnaps=10                      # 保存快照的间隔(每隔多少步保存一次快照)
)
```
